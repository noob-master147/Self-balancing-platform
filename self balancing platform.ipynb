```python
import cv2
import numpy as np
```

```python
cap=cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    blur = cv2.GaussianBlur(frame,(5,5),0)
    ret,th = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    cv2.imshow("image",th)
    
    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
    
    
```
